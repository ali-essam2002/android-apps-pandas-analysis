```python
df=pd.read_csv('google-play-store.csv')
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df2=df.copy()
df2 = df2.dropna(subset=['released'])
```

```python
print(df2.isnull().sum())
```

```python
df2=df2.dropna()
print(df2.isnull().sum())
```

```python
df2['released'] = pd.to_datetime(df['released'])
```

```python
df2.info()
```

### categories of applications get a lot of downloads per day

```python
print(df2.groupby('genre')['ratings_per_day'].mean().sort_values(ascending=False).head(10))
```

### categories of applications don't get many downloads per day

```python
print(df2.groupby('genre')['ratings_per_day'].mean().sort_values().head(10))
```

### what app categories are there market leaders (one app that clearly is getting downloaded more than the others)

```python
print(df.sort_values('min_installs', ascending=False).groupby('genre').first()[['title','min_installs']])
```

### downloads per day might you expect if you took the time to build an app

```python
today = pd.Timestamp.today()

df2['days_since_release'] = (today - df2['released']).dt.days
```

```python
df2['downloads_per_day'] = (df2['min_installs'] / df2['days_since_release'])
```

```python
median_downloads = df2['downloads_per_day'].median()
mean_downloads = df2['downloads_per_day'].mean()

print("Median downloads per day:", round(median_downloads, 2))
print("Mean downloads per day:", round(mean_downloads, 2))
```

### the data tell you about monetization approaches

```python
print(df2.groupby(['price', 'ad_supported', 'offers_iap']).size())
```

### the average rating for each genre

```python
print(df2.groupby('genre')['ratings_per_day'].mean().sort_values(ascending=False).head(10))
```

### top 5 apps with the highest number of reviews

```python
print(df2.groupby("title")['reviews'].sum().sort_values(ascending=False).head(5))
```

### Count the number of apps that offer in-app purchases (IAP)

```python
print("Number of apps offering IAP:", df['offers_iap'].sum())
```

### the total number of installs for each genre

```python
print(df2.groupby('genre')['min_installs'].sum().sort_values(ascending=False).head(10))
```

###  earliest and latest release date in the dataset.

```python
earliest_date = df2['released'].min()
latest_date = df2['released'].max()

print("Earliest release date:", earliest_date)
print("Latest release date:", latest_date)
```

###  average number of ratings per day for apps with more than 10,000 reviews.

```python
print(df2[df2['reviews'] > 10000]['ratings_per_day'].mean())
print(df2[df2['reviews'] > 10000]['ratings_per_day'].median())
```

###  the top 3 genres with the highest average app score.

```python
print(df2.groupby('genre')['score'].mean().sort_values(ascending=False).head(3))
```

###  the percentage of apps that are ad-supported and offer in-app purchases

```python
print((df2['ad_supported'] & df2['offers_iap']).mean() * 100)
```

### most common release year for apps in the dataset.

```python
print("Most common release year:", df2['released'].dt.year.value_counts().idxmax())
print("Number of apps:", df2['released'].dt.year.value_counts().max())
```

### the correlation between the number of reviews and the average rating for each genre

```python
print(df2.groupby('genre')
    .apply(lambda x: x['reviews'].corr(x['score']))
    .sort_values(ascending=False))
```

### the weighted average rating for each genre, where the weight is the number of reviews.

```python
df2['score_x_reviews'] = df2['score'] * df2['reviews']
print(df2.groupby('genre')['score_x_reviews'].sum() / df2.groupby('genre')['reviews'].sum())
```

### dentify the app with the highest ratio of ratings per review for genre.

```python
df2['ratio'] = df2['ratings'] / df2['reviews']
print(df2[df2['reviews'] > 0].groupby('genre')['ratio'].max())
```
